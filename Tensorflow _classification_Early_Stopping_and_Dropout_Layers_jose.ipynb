```python
X = df.drop("benign_0__mal_1",axis=1).values
```

```python
y = df["benign_0__mal_1"].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

###   Train Test Split...test size here taken is very small as the sample size is very small
### Random State is taken arbitrarily

```python
 X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
```

```python
X_train = scale.fit_transform(X_train)
```

## We don't want to fit our test data...we only want to fit into our training data to prevent any leakage

```python
X_test = scale.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
```

### As its a Binary Classification Problem the Output layer will Have Sigmoid as the Activation Function...
### and the loss function will be "binary_crossentropy".....
### Activation Function = Rectified Linear Unit

```python
model.add(Dense(30,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss= "binary_crossentropy",optimizer = "adam")
```

### The epochs have been kept purposely high to induce overfit.....
### Simultaneously Training and validating....

```python
## 600 times through entire dataset
model.fit(x=X_train,y=y_train,epochs=600,validation_data =(X_test,y_test))
```

### Creating Dataframe from Training Loss and The Validation Loss

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses
```

### During Initial Epochs both Training Loss and Validation loss(on Test Dataset) is reducing...
### But as we it moves ahead...The Training Loss is getting reduced but the Validation loss is begining to   increase....
### Basically indicating the overfit

```python
losses.plot()
```

### Redefining the model...with Early stop

```python
model = Sequential()
model.add(Dense(30,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss= "binary_crossentropy",optimizer = "adam")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# help(EarlyStopping)
```

###  mode = "min" =  training will stop when the quantity  monitored has stopped decreasing.....
###  monitor = The quantity that we're monitoring(which is Val_loss in this case)....
###  patience = Number of epochs with no improvement after which training will be stopped...(wait for 25 epochs to check if there's no improvement)


```python
early_stop = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss
```

### The Plot Below shows that the training has stopped (after 86th epoch)...when the difference in loss with training dataset and loss in validation dataset has increased...

```python
model_loss.plot()
```

### To prevent an overfit we can out drop off layer....it turns off percentage of neurons randomly

```python
from tensorflow.keras.layers import Dropout
```

```python
models = Sequential()
models.add(Dense(30,activation="relu"))
models.add(Dropout(0.5))
models.add(Dense(15,activation="relu"))
models.add(Dropout(0.5))
models.add(Dense(1,activation="sigmoid"))
models.compile(loss="binary_crossentropy",optimizer = "adam")
```

### Using Drop off layer with Early stop...both of these things should prevent overfit

```python
early_stop = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
models.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

### It ran little bit longer...due to the fact that there were Drop out Layers....which is actually good

```python
models_loss = pd.DataFrame(models.history.history)
```

```python
models_loss
```

```python
models_loss.plot()
```

### Notice here the Training Loss and the Validation loss are both going down and flattened at the same rate..
### and the Training stopped before it started to Overfit

```python
predictions = models.predict_classes(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
