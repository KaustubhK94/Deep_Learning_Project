### Separating Dependant Feature and the Target Column....

```python
X = df.drop("price",axis=1).values
y = df["price"].values
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

## Performing Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
```

```python
X_train = scale.fit_transform(X_train)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test = scale.transform(X_test)
```

### Importing Dependencies...

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

### Layers and No. of Neurons per layer...with activation Function as ReLU and Adam's Optimizer

```python
optimizer = ["rmsprop","adam",]
model = Sequential()
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(1))
model.compile(optimizer="adam",loss="mse")
```

## Training  and Evaluating The Model Simultaneously

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss
```

## No Spike Indicating there's No Overfitting...

```python
loss.plot()
```

### Evaluation Based on Metrices

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```

```python
df["price"].describe()
```

```python
explained_variance_score(y_test,predictions)
```
